# **Pokeballs**

## 1.a ¿Es cierto que algunas pokebolas son más o menos efectivas dependiendo de propiedades intrínsecas de cada Pokemon? Justificar. *Sugerencia: Comparar efectividad (success/total attemps) como proporción de la efectividad de la Pokebola básica para cada Pokemon*

```python
pokemon = ['caterpie', 'jolteon', 'mewtwo', 'onix', 'snorlax']
pokeball = ['fastball', 'heavyball', 'pokeball', 'ultraball']
```

```python
mean_prob_caught_pokeball = []
error_prob_caught_pokeball = []
pokeballs = []

#for pokeba in pokeball
for j in range(len(pokeball)):
    for i in range(len(pokemon)):
        pokeball_path = 'tp0/data/' + pokemon[i] + '_' + pokeball[j] + '.txt'

        # Leer el archivo 'caterpie.txt'
        with open(pokeball_path, 'r') as archivo:
            # Leer todas las líneas del archivo
            lineas = archivo.readlines()

        pokemon_captured = []
        # Procesar e imprimir cada línea como un numpy array
        for linea in lineas:
            # Convertir la línea de texto en una tupla de Python
            tupla = ast.literal_eval(linea.strip())
            # Convertir la tupla en un numpy array
            array_numpy = np.array(tupla)
            #if array_numpy[0] == 1:
                #pokemon_captured = pokemon_captured + 1
            pokemon_captured.append(array_numpy[0])
        
        #print(pokemon[i], pokeball[j], np.mean(pokemon_captured), np.std(pokemon_captured, ddof=1))

        mean_prob_caught_pokeball.append(np.mean(pokemon_captured))
        #std_prob_caught_pokeball.append(np.std(pokemon_captured, ddof=1))  
        error_prob_caught_pokeball.append(np.var(pokemon_captured, ddof=1))


print(mean_prob_caught_pokeball)
print(error_prob_caught_pokeball)

#print("[[fastball], [heavyball], [pokeball], [ultraball]] =", pokeballs)
# [[fastball], [heavyball], [pokeball], [ultraball]]
```

```python
print(mean_prob_caught_pokeball[0:5])
print(mean_prob_caught_pokeball[5:10])
print(mean_prob_caught_pokeball[10:15])
print(mean_prob_caught_pokeball[15:20])
```

```python
print(error_prob_caught_pokeball[0:5])
print(error_prob_caught_pokeball[5:10])
print(error_prob_caught_pokeball[10:15])
print(error_prob_caught_pokeball[15:20])
```

```python
# Datos de ejemplo: Probabilidades de captura para 4 pokebolas y 5 pokémon
data = {
    "Pokemon": ["Caterpie", "Jolteon", "Mewtwo", "Onix", "Snorlax"],
    "FastBall": mean_prob_caught_pokeball[0:5],
    "Heavyball": mean_prob_caught_pokeball[5:10],
    "Pokeball": mean_prob_caught_pokeball[10:15],
    "Ultraball": mean_prob_caught_pokeball[15:20]
}

# Supongamos que tienes las barras de error (desviaciones estándar) para cada pokebola y pokemon
std_prob_caught_pokeball = {
    "FastBall": error_prob_caught_pokeball[0:5],
    "Heavyball": error_prob_caught_pokeball[5:10],
    "Pokeball": error_prob_caught_pokeball[10:15],
    "Ultraball": error_prob_caught_pokeball[15:20]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Configuración de colores para cada Pokébola
colors = {
    "FastBall": "blue",
    "Heavyball": "green",
    "Pokeball": "orange",
    "Ultraball": "red"
}

# Número de Pokémon y pokebolas
num_pokemon = len(df["Pokemon"])
num_pokebolas = len(df.columns) - 1

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Ancho de cada barra
bar_width = 0.15
# Posiciones para las barras
index = range(num_pokemon)

# Crear barras para cada tipo de pokebola con barras de error
for i, pokebola in enumerate(df.columns[1:]):
    plt.bar(
        [p + i * bar_width for p in index], 
        df[pokebola], 
        bar_width, 
        label=pokebola,
        color=colors[pokebola],
        yerr=std_prob_caught_pokeball[pokebola],  # Añadir barras de error
        capsize=5  # Tamaño de las "tapas" de las barras de error
    )

# Configurar etiquetas y títulos
plt.xlabel('Pokémon')
plt.ylabel('Probabilidad de Captura Promedio')
plt.title('Probabilidad de Captura por Pokébola y Pokémon')
plt.xticks(
    [p + (num_pokebolas / 2) * bar_width for p in index], 
    df["Pokemon"]
)
plt.legend(title='Pokébola', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

## 1b. Ejecutando la función $100$ veces, para cada Pokemon en condiciones ideales (HP:$100$%, LVL $100$) ¿Cuál es la probabilidad de captura promedio para cada pokebola?

```python
pokemon = ['caterpie', 'jolteon', 'mewtwo', 'onix', 'snorlax']
pokeball = ['fastball', 'heavyball', 'pokeball', 'ultraball']
prob_caught_pokeball = []
#for pokeba in pokeball
for j in range(len(pokeball)):
    for i in range(len(pokemon)):
        pokeball_path = 'tp0/data/' + pokemon[i] + '_' + pokeball[j] + '.txt'

        # Leer el archivo 'caterpie.txt'
        with open(pokeball_path, 'r') as archivo:
            # Leer todas las líneas del archivo
            lineas = archivo.readlines()

        pokemon_captured = 0
        # Procesar e imprimir cada línea como un numpy array
        for linea in lineas:
            # Convertir la línea de texto en una tupla de Python
            tupla = ast.literal_eval(linea.strip())
            # Convertir la tupla en un numpy array
            array_numpy = np.array(tupla)
            if array_numpy[0] == 1:
                pokemon_captured = pokemon_captured + 1

        prob_caught_pokeball.append(pokemon_captured/100)

    print(pokeball[j], prob_caught_pokeball)
    prob_caught_pokeball = []
```

```python
# Datos de ejemplo: Probabilidades de captura para 4 pokebolas y 5 pokémon
data = {
    "Pokemon": ["Caterpie", "Jolteon", "Mewtwo", "Onix", "Snorlax"],
    "FastBall": fastball,
    "Heavyball": heavyball,
    "Pokeball": pokeball,
    "Ultraball": ultraball
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Calcular la probabilidad de captura promedio por Pokémon
df['Promedio'] = df.iloc[:, 1:].mean(axis=1)

# Configuración de colores para cada Pokébola
colors = {
    "FastBall": "blue",
    "Heavyball": "green",
    "Pokeball": "orange",
    "Ultraball": "red"
}

# Número de Pokémon
num_pokemon = len(df["Pokemon"])

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Ancho de las barras
bar_width = 0.4  # Ajusta este valor para cambiar el tamaño de las barras

# Crear barras para la probabilidad promedio
plt.bar(
    df["Pokemon"], 
    df['Promedio'], 
    width=bar_width,  # Usa el parámetro 'width' para ajustar el ancho de las barras
    color='purple'
)

# Configurar etiquetas y títulos
plt.xlabel('Pokémon')
plt.ylabel('Efectividad de Captura por Pokémon Promedio')
plt.title('Efectividad de Captura por Pokémon Promedio')
plt.xticks(rotation=45)

# Ajustar el diseño para que se vea bien
plt.tight_layout()
plt.show()
```
